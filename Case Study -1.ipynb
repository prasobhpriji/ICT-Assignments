```python
# Bank Account Manager

account_holder = "Prasobh Priji"
account_number = "98913244083"
account_balance = 1000.00  # Starting balance
```

```python
def show_balance():
    """Function to show current balance"""
    print(f"\nCurrent Balance: ${account_balance:.2f}")

def deposit_money():
    """Function to deposit money"""
    global account_balance

    try:
        amount = float(input("\nEnter amount to deposit: $"))

        if amount > 0:
            account_balance += amount
            print(f"Successfully deposited ${amount:.2f}")
            show_balance()
        else:
            print("Please enter a positive amount!")

    except ValueError:
        print("Please enter a valid number!")

def withdraw_money():
    """Function to withdraw money"""
    global account_balance

    try:
        amount = float(input("\nEnter amount to withdraw: $"))

        if amount > account_balance:
            print("Insufficient funds! You don't have enough money.")
        elif amount <= 0:
            print("Invalid Amount")
        else:
            account_balance -= amount
            print(f"Successfully withdrew ${amount:.2f}")
            show_balance()

    except ValueError:
        print("Please enter a valid number!")

def show_account_info():
    """Function to show account information"""
    print("\n--- Account Information ---")
    print(f"Account Holder: {account_holder}")
    print(f"Account Number: {account_number}")
    show_balance()
```

```python
def main():
    print("=== Welcome to P's Bank ===")

    while True:
        print("\nWhat would you like to do?")
        print("1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. View Account Info")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            show_balance()
        elif choice == "2":
            deposit_money()
        elif choice == "3":
            withdraw_money()
        elif choice == "4":
            show_account_info()
        elif choice == "5":
            print("\nThank you for banking with us! Goodbye!")
            break
        else:
            print("Invalid choice! Please enter a number between 1-5.")

#start
if __name__ == "__main__":
    main()
```

```python
# Weather Data Analyzer

temperature = 25.5  # Celsius
humidity = 70       # Percentage
wind_speed = 15     # Km/h
rainfall = 0        # mm
```

```python
def calculate_average(temp1, temp2, temp3):
    """Function to calculate average temperature"""
    average = (temp1 + temp2 + temp3) / 3
    return average

def calculate_average_humidity(hum1, hum2, hum3):
    """Function to calculate average humidity"""
    average = (hum1 + hum2 + hum3) / 3
    return average

def determine_weather_condition(temp, hum, rain):
    """Function to determine weather condition"""
    if rain > 0:
        return "Rainy"
    elif temp > 30 and hum > 80:
        return "Hot and Humid"
    elif temp > 25:
        return "Sunny"
    elif temp > 15:
        return "Partly Cloudy"
    else:
        return "Cold"

def display_weather_report():
    """Function to display weather report"""
    print("\n=== Weather Report ===")
    print(f"Temperature: {temperature}°C")
    print(f"Humidity: {humidity}%")
    print(f"Wind Speed: {wind_speed} km/h")
    print(f"Rainfall: {rainfall} mm")

    condition = determine_weather_condition(temperature, humidity, rainfall)
    print(f"Weather Condition: {condition}")

def analyze_multiple_days():
    """Function to analyze weather for multiple days"""
    print("\n=== 3-Day Weather Analysis ===")

    temperatures = [22.0, 25.5, 28.0]  # Day 1, Day 2, Day 3
    humidities = [65, 70, 75]          # Day 1, Day 2, Day 3

    avg_temp = calculate_average(temperatures[0], temperatures[1], temperatures[2])
    avg_humidity = calculate_average_humidity(humidities[0], humidities[1], humidities[2])

    print(f"Average Temperature: {avg_temp:.1f}°C")
    print(f"Average Humidity: {avg_humidity:.1f}%")

    print("\nDaily Weather Conditions:")
    for i in range(3):
        condition = determine_weather_condition(temperatures[i], humidities[i], 0)
        print(f"Day {i+1}: {temperatures[i]}°C, {humidities[i]}% - {condition}")
```

```python
def main():
    """Main function to run the weather analyzer"""
    print("=== Weather Data Analyzer ===")

    while True:
        print("\nWhat would you like to do?")
        print("1. View Current Weather")
        print("2. Analyze 3-Day Weather")
        print("3. Exit")

        choice = input("Enter your choice (1-3): ")

        if choice == "1":
            display_weather_report()

        elif choice == "2":
            analyze_multiple_days()

        elif choice == "3":
            print("Thank you for using Weather Analyzer! Goodbye!")
            break

        else:
            print("Invalid choice! Please enter 1-3.")

# Start
if __name__ == "__main__":
    main()
```
